```python
import numpy as np
import pandas as pd
import json
```

# Erste Bereinigung der YELP-Daten

```python
# Datensatz hochladen
with open('../00_Datenbeschaffung/yelp_data_berlin.json') as f:
    yelp_data = json.load(f)
f.close()
```

```python
# JSON-Format normalisieren und als DataFrame umwandeln
yelp_df = pd.json_normalize(yelp_data)
```

```python
yelp_df.shape
```

Durch die API wurden fast 150'000 Elemente gesammelt. Viele davon sind aber wahrscheinlich Duplikate, da YELP sich nicht auf das jeweilige Quartier beschränkt, sondern die 1'000 Geschäfte zurückgibt, welche am nächsten sind.

```python
# Datensatz anschauen
yelp_df.head(10)
```

```python
# Unterschen, was wir für Spalten und was sie für einen Typ haben
yelp_df.dtypes
```

```python
# Duplikate löschen
yelp_df = yelp_df.drop_duplicates(subset=['alias'], keep='first')
```

```python
yelp_df.shape
```

Nachdem die Duplikate gelöscht wurden, gibt es nur noch etwa 19'000 Geschäfte im Datensatz.

```python
# List der unterschiedlichen Kategorien auspacken und in neuen Spalten hinzufügen
cat_df = yelp_df.categories.apply(pd.Series)
```

```python
cat_df
```

```python
yelp_df['category1'] = cat_df[0]
yelp_df['category2'] = cat_df[1]
yelp_df['category3'] = cat_df[2]
```

```python
# Alias:Title Beschreibung auspacken und in neuen Spalten hinzufügen
category1_df = yelp_df.category1.apply(pd.Series)
category2_df = yelp_df.category2.apply(pd.Series)
category3_df = yelp_df.category3.apply(pd.Series)
```

```python
yelp_df['category1'] = category1_df['alias']
yelp_df['category2'] = category2_df['alias']
yelp_df['category3'] = category3_df['alias']
```

```python
yelp_df['title1'] = category1_df['title']
yelp_df['title2'] = category2_df['title']
yelp_df['title3'] = category3_df['title']
```

```python
# Stand des Datensatzes anschauen
yelp_df.head()
```

```python
# Index zurücksetzen
yelp_df.reset_index(drop=True, inplace=True)
```

```python
# Übersicht der neuen Spalten
yelp_df.dtypes
```

```python
# Gibt es gewisse Geschäfte im Datensatz, welche Dauerhaft geschlossen sind?
yelp_df['is_closed'].value_counts()
```

```python
# Reihenfolge der Spalten ändern und nicht notwendige Spalten löschen
yelp_df = yelp_df[['id', 'alias', 'name', 'url', 'review_count', 'rating', 'price', 'location.address1', 'location.address2', 'location.address3', 'location.city', 'location.zip_code', 'location.country', 'location.state', 'location.display_address', 'category1', 'title1', 'category2', 'title2', 'category3', 'title3']]
```

```python
# Übersicht des Datensatzes
yelp_df.head()
```

```python
# Gibt es nur Geschäfte aus Berlin in unserem Datensatz?
yelp_df['location.state'].value_counts()
```

Wir haben auch Geschäfte aus andere Bundesländer... Wir werden das später reinigen. 
Zunächst möchten wir aber die Bezirke im Datensatz hinzufügen. Dies werden wir mit Excel anhand der VLOOKUP-Funktion vornehmen, da wir eine externe Liste der PLZ und Bezirke auf Internet gefunden haben (siehe File "Berlin_Zip_Codes.xls").

Dafür speichern wir den jetzigen Stand des Datensatzes in einem CSV-File.

```python
yelp_df.to_csv('yelp_data_bearbeitet.csv', index = False, encoding='utf-8-sig')
```

# Bereinigung der YELP-Daten nach dem Hinzufügen der Bezirke

Die Bezirke wurden im Datensatz hinzugefüt und in einer neuer Datei gespeichert. 
Wir möchten nun diesen neuen Datensatz hochladen, damit wir die Bereinigung abschliessen können.

```python
# Datensatz hochladen
yelp_b = pd.read_csv('yelp_data_bearbeitet_Bezirke.csv')
```

```python
# Übersicht des Datensatzes
yelp_b.head()
```

```python
# Nur noch die Geschäfte behalten, welche sich im Bundesland "Berlin" befinden.
yelp_b = yelp_b[yelp_b['location.state'] == 'BE']
```

```python
# Haben wir NaN Werte in der Spalte "Bezirk"? Ja! Weil gewisse Geschäfte keine ZIP-Code hatten
yelp_b['location.bezirk'].isna().sum()
```

```python
# Wir entfernen die NaN Werte in der Spalte "Bezirk"
yelp_b = yelp_b[(yelp_b['location.bezirk'].isna()) == False]
```

```python
# Haben wir gewisse Geschäfte ohne Kategorie? Ja!
yelp_b['title1'].isna().sum()
```

```python
# Wir entfernen die NaN Werte in der Spalte "Title1"
yelp_b = yelp_b[(yelp_b['title1'].isna()) == False]
```

```python
# Haben wir NaN Werte in der Spalte Preis? Ja! Es ist aber normal, gewisse Geschäfte (z.B. Spitäler, Schulen) haben keine Preiskategorie.
# Wir entfernen diese NaN Werte also nicht.
yelp_b['price'].isna().sum()
```

```python
# Nach der Bereinigung haben wir noch etwa 17'900 Geschäfte in unserem Datensatz.
yelp_b.shape
```

```python
# Datensatz als CSV-Datei exportieren, damit wir mit der Visualisierung weitermachen können.
yelp_b.to_csv('yelp_data_cleaned.csv', index = False, encoding='utf-8-sig')
```
